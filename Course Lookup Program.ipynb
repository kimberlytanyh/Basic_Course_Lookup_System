```python
# Create a program that does the following:

# Read the file here to create a dictionary where the course code is the key 
# and the course name and credit amount are the corresponding values.

# Get user input to do either of the following tasks: 
# (at this point you should be working with the dictionary created from the file, not the file itself)

# Display all the information in the dictionary in a tabular format. 
# Each element should be a row with columns for course code, course name and credit amount

# Allow the user to search the dictionary by course code. 
# Once a match is found, display the course code, course name and credit amount. 
# If a match is not found, display "This course could not be found" or a similar message.

# Note: The user should only be able to do one task at a time so you'll need to provide them 
# with a way to input their choice and a way for them to exit the program.
# Ensure that the user does not enter input that can potentially throw an exception. 
# You must include at least one try/except clause wherever necessary.
```

```python
readFile = open('accounting.txt','r')

course_dictionary = {}

line = readFile.readline().rstrip()

while line != "":
    course_code = line
    course_name = readFile.readline().rstrip()
    credit_no = readFile.readline().rstrip()
    course_dictionary[course_code] = [course_name, credit_no]
    
    line = readFile.readline().rstrip()

readFile.close()

def instructions():
    print("\nPlease input what you would like to do.")
    print("Enter '1' if you want to view all courses.")
    print("Enter '2' if you want to search for a course by course code.\n")

# Get user input to do either of the following tasks:
def user_input():
    keep_asking = True

    while keep_asking:
        try:
            which = int(input("What would you like to do? (1 or 2) "))
            keep_asking = False
            return which
        except ValueError:
            print("Please enter either '1' or '2' for desired action to execute.")
            keep_asking = True
            
instructions()            
which = user_input()

while which == 1 or which == 2:
    if which == 1:
        print("\n\nCourse Code:", "Credit Amount:", "Course Name:", sep = "\t", end = "\n")
        for key in course_dictionary:
            values = course_dictionary[key]
            course_name = values[0]
            credit_number = values[1]
            print(key, credit_number, course_name, sep = "\t", end = "\n")
        other = input("\nWould you like to do something else? Y/N)")
        if other == "Y":
            instructions()
            which = user_input()
        elif other == "N":
            print("Thank you and have a great day.")
            break

    if which == 2:
        search = input("\nEnter the course code of the course you are searching for: ")
        if search in course_dictionary:
            values = course_dictionary[search]
            course_name = values[0]
            credit_number = values[1]
            print("\nOne match found: ")
            print("\n" + search, course_name, credit_number, sep = "\t")
        else:
            print("This course could not be found.")
        search_again = input("\nWould you like to do search another course?(Y/N) ")
        if search_again == "Y":
            which = 2
        elif search_again == "N":
            other = input("Would you like to do something else?(Y/N) ")
            if other == "Y":
                instructions()
                which = user_input()
            elif other == "N":
                print("\nThank you and have a good day.")
                break
```
